```python
import sqlite3
import pandas as pd
import os

# Specify the SQLite database file
database_file = 'data2.db'

# Specify the root directory containing CSV files and subdirectories
csv_root_directory = '../csv2/'

# Function to create tables and insert data from CSV files into the SQLite database
def create_database_from_csv(csv_directory, db_connection):
    cursor = db_connection.cursor()
    
    for root, _, files in os.walk(csv_directory):
        for csv_file in files:
            if csv_file.endswith('.csv'):
                csv_file_path = os.path.join(root, csv_file)
                
                try:
                    # Read the CSV file into a DataFrame
                    df = pd.read_csv(csv_file_path)
                    
                    # Get the table name (use the file name without the extension)
                    table_name = os.path.splitext(csv_file)[0]
                    
                    if not df.empty:
                        # Create a table with the same structure as the CSV file
                        df.to_sql(table_name, db_connection, if_exists='replace', index=False)
                        
                        print(f"Created table '{table_name}' from '{csv_file_path}'")
                    else:
                        print(f"Skipped '{csv_file_path}' because it has no columns to parse.")
                except pd.errors.EmptyDataError:
                    print(f"Skipped '{csv_file_path}' because it has no columns to parse.")
    
    cursor.close()

# Create or connect to the SQLite database
conn = sqlite3.connect(database_file)

# Call the function to create tables and insert data from CSV files
create_database_from_csv(csv_root_directory, conn)

# Commit the changes and close the database connection
conn.commit()
conn.close()

# Function to query data from the SQLite database
def query_database(database_file, query):
    conn = sqlite3.connect(database_file)
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df
```

```python

# Function to get the schema of all tables in the database
def get_database_schema(database_file):
    conn = sqlite3.connect(database_file)
    cursor = conn.cursor()
    
    # Get the names of all tables in the database
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    table_names = cursor.fetchall()
    
    # Get the schema of each table
    schema_dict = {}
    for table_name in table_names:
        cursor.execute(f"PRAGMA table_info(\"{table_name[0]}\")")
        schema = cursor.fetchall()
        schema_dict[table_name[0]] = schema
    
    cursor.close()
    conn.close()
    
    return schema_dict

# Call the function to get the schema of all tables in the database
database_file = 'data2.db'
schema_dict = get_database_schema(database_file)

# Print the schema of each table
for table_name, schema in schema_dict.items():
    print(f"Schema of table '{table_name}':")
    for column in schema:
        print(f"- {column[1]} ({column[2]})")
    print()
```

```python
# Code to search converation between 2 users

import os
import pandas as pd
import sqlite3

# Define a function to get the list of files that contain the user_id
def get_files_with_user_id(user_id, directory):
    files_with_user_id = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                try:
                    df = pd.read_csv(file_path)
                    if 'user' in df.columns and user_id in df['user'].unique():
                        files_with_user_id.append(file_path)
                    else:
                        print(f"Skipped '{file_path}' because it has no 'user' column or '{username}' is not in the 'user' column.")
                except pd.errors.EmptyDataError:
                    print(f"Skipped '{file_path}' because it has no columns to parse.")
    return files_with_user_id

# Define a function to get the DataFrame of all messages sent by the user
def get_messages_by_users(files, user1_id, user2_id, conn):
    cursor = conn.cursor()
    cursor.execute(f"SELECT name FROM users WHERE id='{user1_id}'")
    user1_name = cursor.fetchone()[0]
    cursor.execute(f"SELECT name FROM users WHERE id='{user2_id}'")
    user2_name = cursor.fetchone()[0]
    messages = []
    for file in files:
        try:
            df = pd.read_csv(file)
            if 'type' in df.columns and 'user' in df.columns and 'text' in df.columns:
                if df['type'].iloc[0] == 'message' and user1_id in df['user'].unique() and user2_id in df['user'].unique():
                    messages.extend(df[(df['user'] == user1_id) | (df['user'] == user2_id)]['text'].tolist())
                else:
                    print(f"Skipped '{file}' because it is not a message file or '{user1_id}' or '{user2_id}' is not in the 'user_id' column.")
            else:
                print(f"Skipped '{file}' because it has missing columns.")
        except pd.errors.EmptyDataError:
            print(f"Skipped '{file}' because it has no columns to parse.")
    return pd.DataFrame({'user1': [user1_name]*len(messages), 'user2': [user2_name]*len(messages), 'text': messages})

# Create or connect to the SQLite database
database_file = 'data2.db'
conn = sqlite3.connect(database_file)

# Get the user_id from the users table
user1_name = input("Enter the username of user 1: ")
cursor = conn.cursor()
cursor.execute(f"SELECT id FROM users WHERE name='{user1_name}'")
user1_id = cursor.fetchone()[0]

user2_name = input("Enter the username of user 2: ")
cursor.execute(f"SELECT id FROM users WHERE name='{user2_name}'")
user2_id = cursor.fetchone()[0]

# Call the functions to get the DataFrame of all messages sent by the users
directory = '../csv2/'
files_with_user1_id = get_files_with_user_id(user1_id, directory)
messages_by_users = get_messages_by_users(files_with_user1_id, user1_id, user2_id, conn)
print(messages_by_users)

# Close the database connection
conn.close()
```
